```python
df_continentcases = df.groupby('Continent')['Total Cases'].sum().reset_index()
plt.bar(df_continentcases['Continent'], df_continentcases['Total Cases'], color='pink')
plt.title('Continent-wise Cases')
plt.ylabel('Total Cases')
plt.xlabel('Continent')
plt.xticks(rotation=45, ha='right')
plt.show()
#Plottting the graph for continent wise case.
```

Europe had the most number of cases followed by Asia then North America.
While Oceania had the least number of cases.

```python
#Finding relationship between Population and Total Cases
sns.scatterplot(data=df, x='Population', y='Total Cases', color= 'black')
plt.title('Relationship between Population and Total Cases')
plt.xlabel('Population')
plt.ylabel('Total Cases')
plt.show()
```

There are 3 outliners.

1st one shows that the total no. of cases and population is less which means that most of the people are infected of that place.

2nd and 3rd one shows that the population is high but there is very less no. of cases which means despite the huge population there are very less people infected.
